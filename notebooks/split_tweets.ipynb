```python
for pod in pod_csvs:
    print('Working on {}...'.format(pod[:-4]), end='')
    if not os.path.exists(os.path.join(data_dir, pod[:-4])):
        os.mkdir(os.path.join(data_dir, pod[:-4]))
    
    all_data = pd.read_csv(os.path.join(data_dir, pod))
    tweet_data = all_data.loc[all_data['tweets'] != '[]']
    
    for x2 in tweet_data.index.values.tolist():
        out_file_path = os.path.join(data_dir, pod[:-4], (str(tweet_data.loc[x2, 'user_id']) + '.txt'))
        out_file = open(out_file_path,'w+')
        out_file.write(tweet_data.loc[x2, 'tweets'])
        out_file.close()
    print('Done')    
```
